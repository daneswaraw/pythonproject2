```python
plt.figure(figsize=(8,6))
plt.plot(iris['latitude'], label='LT')
plt.plot(iris['longitude'], label='LG')
plt.plot(iris['h_ellipsoid_m'], label='HEM')
           
plt.xlabel('NUMBER OF DATA')
plt.ylabel('VALUE')
plt.title('Data Penurunan Tanah di DKI Jakarta Tahun 2014')
plt.legend()
plt.show()
```
